## Genocídio

* matar todo mundo SEM PERDER O MELHOR DE TODOS, duas opções:
    * matar todos menos o melhor de todos e reiniciar o processo
    * matar todos incluindo o melhor de todos (guardando ele em uma variável global); reiniciar o processo e comparar se o novo melhor de todos é melhor ou pior que o antigo; caso seja pior, podemos re-inserir o melhor de todos antigo

- o genocído é muito interessante quando é fácil convergir para um máximo local e dificl de achar o global 
    - genocidio é muito caro computacionalmente, podemos melhorar isso de alguma forma:

## Eugenia

- nesse mesmo pique de matar a opulação, podemos fazer uma eugenia: matar os piores
    - interessante para o caso de realizarmos um genocidio com uma pop maior do que o que queremos; ai assim podemos realizar um eugenia para tirar os individus a mais da população; cuidado com a diversidade!

* o genocidio aumenta absurdamente a diversidade, a eugenia diminui um pouco ela; a ideia é, se ja usamos o genocidio é por que trancamos e queremos procurar outro máximo local melhor ou um global; se perdemos a mão na eugenia o genocidio talvez não tenha valido tanto a pena
